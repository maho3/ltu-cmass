## Port Abacus
Ports Abacus cosmologies to our parameter definitions and saves them to the parameter file.

### Old way of doing it (Before 5/2/25)
Before we started customizing the order of simulations

```python
import numpy as np
import pandas as pd

cosm = pd.read_csv(
    '/anvil/scratch/x-mho1/cmass-ili/scratch/raw_abacus_cosmologies.csv',
    skipinitialspace=True, delimiter=',')
cosm.columns = [c.strip() for c in cosm.columns]
index = cosm['root'].apply(lambda x: int(x.strip()[-3:]))
index.name = 'index'
cosm.index = index

cosm.loc[19, 'omega_ncdm'] = cosm.loc[19, 'omega_ncdm'].split(',')[
    0]  # fix this bad line
cosm['omega_ncdm'] = cosm['omega_ncdm'].values.astype(float)
cosm
```

```python
Omega_m = cosm['omega_cdm'] + cosm['omega_b'] + cosm['omega_ncdm']
Omega_b = cosm['omega_b']
h = cosm['h']
n_s = cosm['n_s']
sigma8 = cosm['sigma8_m']

# rescale to capital Omega
Omega_m /= h**2
Omega_b /= h**2

out = np.zeros((max(index)+1, 5))
out[index, 0] = Omega_m
out[index, 1] = Omega_b
out[index, 2] = h
out[index, 3] = n_s
out[index, 4] = sigma8

out
```

```python
np.savetxt(
    '/home/x-mho1/git/ltu-cmass/params/abacus_cosmologies.txt',
    out, delimiter=' ', fmt='%f')
```
